### 1- Definição da função objetivo

#### A função fornecida é a função de Rosenbrock, também conhecida como "função da banana" ou "vale de Rosenbrock". Esta função é comumente usada para testar algoritmos de otimização devido à sua forma característica. Aqui está uma breve descrição da função:$f(x,y)=100(y−x^2)^2 +(1−x)^2$


image.png

 #### Ela é não-convexa e possui um mínimo global em $f(1,1)=0$. A função é comumente usada para testar algoritmos de otimização, pois apresenta um desafio devido à sua forma estreita e longa, criando uma trajetória de otimização não trivial para muitos métodos.

```python
def rosenbrock(x, y):
    """
    Função de Rosenbrock, também conhecida como "função da banana" ou "vale de Rosenbrock".

    Parâmetros:
    x (float): Valor da variável x.
    y (float): Valor da variável y.

    Retorna:
    float: Valor da função de Rosenbrock para as variáveis x e y.
    """
    return 100 * (y - x**2)**2 + (1 - x)**2
```

### 2- Obtenção das Curvas de Nível

```python
# Cria um vetor com 100 pontos uniformemente espaçados no intervalo de -3 a 3
_ = np.linspace(-3, 3, 100)

# Cria uma grade bidimensional usando np.meshgrid com base no vetor _
# As coordenadas resultantes são atribuídas às variáveis x e y
x, y = np.meshgrid(_, _)

# Calcula os valores da função de Rosenbrock para cada par de coordenadas (x, y) na grade
z = 100*(y - x**2)**2 + (1 - x)**2

# Cria um gráfico de contorno preenchido usando os valores calculados e as coordenadas da grade
# contourf é usado para criar áreas preenchidas, indicando diferentes níveis de altura da função de Rosenbrock
plt.contourf(x, y, z)
```

### 5 - Algoritmo de recozimento simulado

```python
# Define uma semente para a geração de números aleatórios, garantindo reprodutibilidade
random.seed(42)

# Função para executar o recozimento simulado
def simulated_annealing(energy_func, initial_state, iterations, initial_temperature, cooling_rate, lb, ub, attempts):
    # Lista para armazenar as melhores energias encontradas em cada iteração
    record_best_energy = []

    # Lista para armazenar as melhores soluções encontradas em cada iteração
    best_sol = []

    # Inicializa o estado atual com as coordenadas iniciais
    current_state = initial_state

    # Calcula a energia inicial usando a função de energia fornecida
    current_energy = energy_func(*current_state)

    # Inicializa as melhores coordenadas e energia com os valores iniciais
    best_state = current_state
    best_energy = current_energy

    # Imprime informações sobre os parâmetros iniciais do algoritmo
    print(f'Starting Simulated Annealing with {iterations} iterations, {initial_temperature} initial temperature, {cooling_rate} cooling rate, and [{lb}, {ub}] limits')

    # Loop principal para iterar sobre o número total de iterações
    for i in range(iterations):
        # Loop interno para tentativas em cada iteração
        for j in range(attempts):
            # Atualiza a temperatura de acordo com a taxa de resfriamento
            temperature = initial_temperature / (1 + cooling_rate * i)

            # Gera um novo estado perturbando o estado atual dentro dos limites definidos
            new_state = (current_state[0] + random.uniform(lb, ub), current_state[1] + random.uniform(lb, ub))

            # Calcula a energia do novo estado usando a função de energia fornecida
            new_energy = energy_func(*new_state)

            # Calcula a diferença de energia entre os estados
            energy_delta = new_energy - current_energy

            # Verifica se o novo estado tem menor energia ou segue a probabilidade de Boltzmann
            if energy_delta < 0 or random.uniform(0, 1) < math.exp(-energy_delta / temperature):
                # Atualiza o estado atual se as condições forem atendidas
                current_state = new_state
                current_energy = new_energy

                # Atualiza as melhores coordenadas se a nova energia for menor que a melhor energia
                if new_energy < best_energy:
                    best_state = new_state
                    best_energy = new_energy

        # Salva o último valor de solução e energia de cada tentativa
        best_sol.append(best_state)
        record_best_energy.append(best_energy)

        # Imprime os valores apenas a cada 1000 iterações
        if i % 1000 == 0:
            print("Iteration: {}, Best Solution: {}, Best Fitness: {}".format(i, best_state, best_energy))

    # Gráfico de convergência mostrando a evolução da melhor energia ao longo das iterações
    plt.plot(record_best_energy, '-k')

    # Retorna o valor mínimo da energia salva e a última solução encontrada
    return min(record_best_energy), best_sol[-1]

# Função de Rosenbrock (como exemplo, substitua por sua própria função)
def rosenbrock(x, y):
    return 100 * (y - x**2)**2 + (1 - x)**2

# Parâmetros iniciais para o algoritmo
initial_state = (0, 0)
iterations = 10000
initial_temperature = 100.0
cooling_rate = 0.1
lb = -2
ub = 2
attempts = 100

# Executa o recozimento simulado para encontrar o mínimo da função de Rosenbrock
min_energy, final_solution = simulated_annealing(rosenbrock, initial_state, iterations, initial_temperature, cooling_rate, ub, lb, attempts)

# Imprime o resultado final
print("Minimum Energy Found: {}".format(min_energy))
print("Final Solution: {}".format(final_solution))

# Mostra o gráfico de convergência
plt.title("Convergence Plot")
plt.xlabel("Iteration")
plt.ylabel("Best Energy")
plt.show()
```
